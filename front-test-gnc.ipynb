### Ajustar valores da altura em cm.

```python
# a partir de uma coluna que existe
tabela_gnc_df['altura'] = tabela_gnc_df['altura']*10
display(tabela_gnc_df)

```

### Disparando eventos

    - Botão para mostrar 10 valores da planilha.
        
        * Nome
        * Altura
        * Peso
        * Aniversário
        

```python
#import ipwidgets as widgets
wid_button = widgets.Button(description="Mostrar Planilha")
wid_output = widgets.Output()
def on_button_clicked(b):
    with wid_output:
        tabela_gnc_df = pd.read_excel("tabela_gnc.xlsx")
        select = tabela_gnc_df[['nome', 'altura', 'peso', 'aniversario',]]
        display(select.loc[1:10])
    
wid_button.on_click(on_button_clicked)
display(wid_button, wid_output)
```
